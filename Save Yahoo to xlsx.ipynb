# Save yahoo data to xlsx

```python
from pandas_datareader import data, wb
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = dt.datetime(2013,1,1)
end = dt.datetime(2017,4,20)
#Search stock code and data
data = data.DataReader('0700.HK','yahoo', start, end)
```

```python
data.plot()
```

```python
data.tail()
```

```python
data['Adj Close'].plot()
```

```python
data.to_excel('Stock_Price_0700HK.xlsx')
```
